```python
hd["ts_parsed"] = pd.to_datetime(hd["Timestamp IST"], format="%d-%m-%Y %H:%M")
print("Dataset 1:", hd["ts_parsed"].min(), "to", hd["ts_parsed"].max())

fg["ts_parsed"] = pd.to_datetime(fg["timestamp"], unit="s")
print("Dataset 2:", fg["ts_parsed"].min(), "to", fg["ts_parsed"].max())
```

```python
hd["date_only"] = hd["ts_parsed"].dt.date
fg["date_only"] = fg["ts_parsed"].dt.date

start_overlap = max(hd["ts_parsed"].min(), fg["ts_parsed"].min())
end_overlap   = min(hd["ts_parsed"].max(), fg["ts_parsed"].max())

hd_filtered = hd[(hd["ts_parsed"] >= start_overlap) & (hd["ts_parsed"] <= end_overlap)]
fg_filtered = fg[(fg["ts_parsed"] >= start_overlap) & (fg["ts_parsed"] <= end_overlap)]

df = pd.merge(hd_filtered, fg_filtered, on="date_only", how="inner")

print(df.shape)
print(df.head())
```

```python
df = df.drop(columns=[
    "Transaction Hash",
    "Order ID",
    "Trade ID",
    "Crossed",
    "Timestamp IST",
    "Timestamp",
    "timestamp",
    "date",
    "ts_parsed_y",
])
```

Day PnL = sum of all Closed PnL values for that account on that day

Win Rate = (number of trades where Closed PnL > 0) / (total trades) × 100

Avg Trade Size = mean of Size USD column

L/S Ratio = number of Long trades / number of Short trades

Leverage is the amount of borrowed money it ampifies both loss and profit so it is expected it will be low in fear days

```python
df["is_win"] = df["Closed PnL"] > 0
df["approx_leverage"] = (
    df["Size USD"] / df["Start Position"].abs().replace(0, np.nan)
).round(2)

daily_pnl = (
    df.groupby(["date_only", "Account"])["Closed PnL"]
    .sum()
    .reset_index()
    .rename(columns={"Closed PnL": "daily_pnl"})
)
df = df.merge(daily_pnl, on=["date_only", "Account"], how="left")

win_rate = (
    df.groupby("Account")["is_win"]
    .mean()
    .reset_index()
    .rename(columns={"is_win": "win_rate_%"})
)
win_rate["win_rate_%"] = (win_rate["win_rate_%"] * 100).round(2)
df = df.merge(win_rate, on="Account", how="left")

avg_trade_size = (
    df.groupby("Account")["Size USD"]
    .mean()
    .reset_index()
    .rename(columns={"Size USD": "avg_trade_size_usd"})
)
df = df.merge(avg_trade_size, on="Account", how="left")

trades_per_day = (
    df.groupby("date_only")
    .size()
    .reset_index(name="num_trades")
)
df = df.merge(trades_per_day, on="date_only", how="left")

long_short = (
    df.groupby(["date_only", "Direction"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)
if "Long" in long_short.columns and "Short" in long_short.columns:
    long_short["long_short_ratio"] = (
        long_short["Long"] / long_short["Short"].replace(0, np.nan)
    ).round(2)
    df = df.merge(long_short[["date_only", "long_short_ratio"]], on="date_only", how="left")

print(df.columns.tolist())
print(df.shape)
```

Fix the NaN Leverage

```python
df["approx_leverage"] = (
    df["Size USD"] / df["Start Position"].abs().replace(0, np.nan)
).round(2).fillna(1.0)
```

```python
df.head()
```

```python
df["is_drawdown"] = df["Closed PnL"] < 0
```

```python
sentiment_perf = df.groupby("classification").agg(
    avg_pnl        = ("Closed PnL", "mean"),
    total_pnl      = ("Closed PnL", "sum"),
    win_rate       = ("is_win", "mean"),
    drawdown_rate  = ("is_drawdown", "mean"),
    num_trades     = ("Closed PnL", "count")
).reset_index()

sentiment_perf["win_rate"]      = (sentiment_perf["win_rate"] * 100).round(2)
sentiment_perf["drawdown_rate"] = (sentiment_perf["drawdown_rate"] * 100).round(2)

print(sentiment_perf)
```

```python
order = ["Extreme Fear", "Fear", "Neutral", "Greed", "Extreme Greed"]
sentiment_perf["classification"] = pd.Categorical(
    sentiment_perf["classification"], categories=order, ordered=True
)
sentiment_perf = sentiment_perf.sort_values("classification")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Performance by Fear & Greed Sentiment", fontsize=15, fontweight="bold")

colors = ["#d73027", "#f46d43", "#ffffbf", "#74add1", "#313695"]

axes[0].bar(sentiment_perf["classification"], sentiment_perf["avg_pnl"], color=colors)
axes[0].set_title("Average PnL per Trade")
axes[0].set_xlabel("Sentiment")
axes[0].set_ylabel("Avg Closed PnL (USD)")
axes[0].axhline(0, color="black", linewidth=0.8, linestyle="--")
axes[0].tick_params(axis="x", rotation=30)

axes[1].bar(sentiment_perf["classification"], sentiment_perf["win_rate"], color=colors)
axes[1].set_title("Win Rate (%)")
axes[1].set_xlabel("Sentiment")
axes[1].set_ylabel("Win Rate %")
axes[1].tick_params(axis="x", rotation=30)

axes[2].bar(sentiment_perf["classification"], sentiment_perf["drawdown_rate"], color=colors)
axes[2].set_title("Drawdown Rate (% losing trades)")
axes[2].set_xlabel("Sentiment")
axes[2].set_ylabel("% Losing Trades")
axes[2].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.savefig("section1_performance.png", dpi=150)
plt.show()
```

```python
sentiment_behavior = df.groupby("classification").agg(
    avg_leverage       = ("approx_leverage", "mean"),
    avg_trade_size_usd = ("Size USD", "mean"),
    avg_num_trades     = ("num_trades", "mean"),
).reset_index()

sentiment_behavior["classification"] = pd.Categorical(
    sentiment_behavior["classification"], categories=order, ordered=True
)
sentiment_behavior = sentiment_behavior.sort_values("classification")
```

```python
ls_sentiment = df.groupby(["classification", "Direction"]).size().unstack(fill_value=0)
ls_sentiment["long_short_ratio"] = (
    ls_sentiment["Buy"] / ls_sentiment["Sell"].replace(0, np.nan)
).round(2)
ls_sentiment = ls_sentiment.reset_index()
ls_sentiment["classification"] = pd.Categorical(
    ls_sentiment["classification"], categories=order, ordered=True
)
ls_sentiment = ls_sentiment.sort_values("classification")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Trader Behavior by Fear & Greed Sentiment", fontsize=15, fontweight="bold")

axes[0].bar(sentiment_behavior["classification"], sentiment_behavior["avg_leverage"], color=colors)
axes[0].set_title("Avg Leverage")
axes[0].set_xlabel("Sentiment")
axes[0].set_ylabel("Leverage (x)")
axes[0].tick_params(axis="x", rotation=30)

axes[1].bar(sentiment_behavior["classification"], sentiment_behavior["avg_trade_size_usd"], color=colors)
axes[1].set_title("Avg Trade Size (USD)")
axes[1].set_xlabel("Sentiment")
axes[1].set_ylabel("Size USD")
axes[1].tick_params(axis="x", rotation=30)

axes[2].bar(ls_sentiment["classification"], ls_sentiment["long_short_ratio"], color=colors)
axes[2].set_title("Long/Short Ratio")
axes[2].set_xlabel("Sentiment")
axes[2].set_ylabel("Ratio (>1 = more longs)")
axes[2].axhline(1, color="black", linewidth=0.8, linestyle="--")
axes[2].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.savefig("section2_behavior.png", dpi=150)
plt.show()
```

```python
account_stats = df.groupby("Account").agg(
    total_pnl      = ("Closed PnL", "sum"),
    avg_pnl        = ("Closed PnL", "mean"),
    win_rate       = ("is_win", "mean"),
    avg_leverage   = ("approx_leverage", "mean"),
    trade_count    = ("Closed PnL", "count"),
    avg_size_usd   = ("Size USD", "mean"),
).reset_index()

account_stats["win_rate"] = (account_stats["win_rate"] * 100).round(2)
```

```python
lev_median = account_stats["avg_leverage"].median()
account_stats["leverage_segment"] = np.where(
    account_stats["avg_leverage"] >= lev_median, "High Leverage", "Low Leverage"
)
```

```python
trade_median = account_stats["trade_count"].median()
account_stats["frequency_segment"] = np.where(
    account_stats["trade_count"] >= trade_median, "Frequent", "Infrequent"
)
```

```python
account_stats["consistency_segment"] = np.where(
    (account_stats["win_rate"] >= 50) & (account_stats["total_pnl"] > 0),
    "Consistent Winner", "Inconsistent"
)

print(account_stats[["leverage_segment","frequency_segment","consistency_segment"]].value_counts())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Trader Segmentation Analysis", fontsize=15, fontweight="bold")

seg1 = account_stats.groupby("leverage_segment")[["win_rate","avg_pnl"]].mean().reset_index()
axes[0].bar(seg1["leverage_segment"], seg1["win_rate"], color=["#d73027","#313695"])
axes[0].set_title("Win Rate: High vs Low Leverage")
axes[0].set_ylabel("Win Rate %")

seg2 = account_stats.groupby("frequency_segment")[["total_pnl","avg_leverage"]].mean().reset_index()
axes[1].bar(seg2["frequency_segment"], seg2["total_pnl"], color=["#f46d43","#74add1"])
axes[1].set_title("Total PnL: Frequent vs Infrequent")
axes[1].set_ylabel("Avg Total PnL (USD)")
axes[1].axhline(0, color="black", linewidth=0.8, linestyle="--")

seg3 = account_stats["consistency_segment"].value_counts().reset_index()
axes[2].pie(
    seg3["count"],
    labels=seg3["consistency_segment"],
    autopct="%1.1f%%",
    colors=["#74add1","#f46d43"]
)
axes[2].set_title("Consistent Winners vs Inconsistent")

plt.tight_layout()
plt.savefig("section3_segments.png", dpi=150)
plt.show()
```



1.  **Sentiment's Impact on Performance**: The analysis shows that traders tend to perform the best (in terms of both average profit and win rate) when sentiment is "Extreme Greed." Traders under "Extreme Fear" tend to perform the worst, with lower average profits and win rates. This suggests that market sentiment, especially extreme sentiment, plays a significant role in trading performance.


2.  **Behavior Differences by Sentiment**: Traders in the "Extreme Fear" sentiment group tend to take lower leverage and smaller trade sizes, while those in "Extreme Greed" exhibit higher leverage and trade sizes. This suggests that traders' behavior, in terms of leverage and trade sizes, is heavily influenced by their sentiment, with more aggressive strategies (high leverage and trade sizes) used when sentiment is more greedy.


3.  **Consistency in Trading**: The segmentation analysis reveals that the majority of traders (90.6%) are "Inconsistent Winners," suggesting that most traders don't achieve consistent profitability over time. The small percentage of "Consistent Winners" highlights the challenge of maintaining steady success in trading, which could be attributed to various factors such as emotional biases, risk management, and market conditions.


# Strategy 1: During Extreme Greed, Increase Trade Size and Leverage for High-Leverage Traders
# Strategy 2: During Extreme Fear, Reduce Leverage and Limit Trade Size for Inconsistent Winners

# Trading Strategy Ideas Based on Sentiment Analysis

### 1. **During Extreme Greed, Increase Trade Size and Leverage for High-Leverage Traders**
- **Rule**: Traders with high leverage should increase their trade size and leverage during periods of "Extreme Greed."
- **Why**: The analysis showed that traders using higher leverage tend to perform better in "Extreme Greed" sentiment. This strategy takes advantage of more favorable market conditions by amplifying positions.

---

### 2. **During Extreme Fear, Reduce Leverage and Limit Trade Size for Inconsistent Winners**
- **Rule**: Traders with low win rates (Inconsistent Winners) should reduce their leverage and limit their trade size during periods of "Extreme Fear."
- **Why**: During "Extreme Fear," traders with a lower win rate tend to suffer from more significant drawdowns. By reducing their leverage and trade size, they can minimize risk and avoid larger losses during these unfavorable market conditions.
